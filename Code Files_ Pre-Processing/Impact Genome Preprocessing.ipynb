```python
import pandas as pd
import numpy as np
```

```python
fileName = 'Validated Data from Heather.xlsx'
full_df = pd.read_excel(fileName)
```

```python
full_df.head(5)
```

```python
#Drop all null values
full_df = full_df.dropna()
full_df.count()
```

```python
#Convert outcomeid to integer type
full_df['outcomeid'] = full_df['outcomeid'].astype(int)
full_df.dtypes
```

```python
#Remove special characters from program description
import re
full_df['programdescription'] = [re.sub(r"[^A-Za-z0-9 :.,'-]+", "", x) for x in full_df['programdescription']]
#Leave only letters, numbers, and some punctutation marks (colon, period, comma, apostraphe, hyphen)
```
